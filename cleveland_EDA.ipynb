```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"
]
```

```python
df = pd.read_csv("/content/processed.cleveland (1).data", names=column_names, na_values="?")
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="target", data=df)
plt.title("Target Class Distribution")
plt.xlabel("Heart Disease (0 = No, 1â€“4 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
df.hist(bins=20, figsize=(15, 12), edgecolor='black')
plt.suptitle("Histogram of All Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
for col in ["age", "trestbps", "chol", "thalach", "oldpeak"]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x="target", y=col, data=df)
    plt.title(f"{col} vs Target")
    plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
selected_cols = ["age", "thalach", "oldpeak", "ca", "thal", "target"]
sns.pairplot(df[selected_cols], hue="target", corner=True)
plt.show()
```
